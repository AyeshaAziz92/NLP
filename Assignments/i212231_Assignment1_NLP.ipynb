```python
import requests
from bs4 import BeautifulSoup
import re 
# Make the GET request to a url
res = requests.get('https://en.wikipedia.org/wiki/Abraham_Lincoln').text
```

# **a) Years mentioned in a range like 1998-2000**

```python
# Extracting years mentioned in a range like 1998-2000.
years = re.findall('1998|1999|2000',res)
print(years)
```

# **b) words beginning with a capital letter.**

```python
# Extracting words beginning with a capital letter
letters = re.findall('([A-Z].*?)[ |.|\\\n,"></;={}%#$]',res)
print(letters)
```

```python
soup = BeautifulSoup(res,'lxml')
print(set([text.parent.name for text in soup.find_all(text=True)]))
```

# **c) heading and sub-headings**

```python
# Extracting Headings and sub headings 
heading_tags = ["h1", "h2","h3"]
data = []
for tags in soup.find_all(heading_tags):
    x = tags.text.strip()
    data.append(x)
print(data)
```

**Mereged all data into 1 list**

```python
new = letters + years + data
print(new)
```

# **2) convert the complete text to lower case**

```python
import nltk
# convert the complete text to lower case
lower =[]
for x in new:
  lower_text = x.lower()
  lower.append(lower_text)
print(lower)
type(list)
```

# **3) Remove all the stop words**

```python
#removing stopword

import nltk
from nltk.corpus import stopwords
stopword = stopwords.words('english')
new_list=[]
for x in lower:
  word_tokens = nltk.word_tokenize(x)
  removing_stopwords = [word for word in word_tokens if word not in stopword]
  new_list.append(removing_stopwords)
print (new_list)
```

```python
listOfList = new_list
# Use list comprehension to convert a list of lists to a flat list 
flatList = [ item for elem in listOfList for item in elem]
print( flatList)   
```

# **5) Apply POS tagging on the whole text**

```python
tokenized = sent_tokenize(listToStr)
for i in tokenized:
      
    wordsList = nltk.word_tokenize(i)
  
    # removing stop words from wordList
    wordsList = [w for w in wordsList if not w in stop_words] 
  
     
    POStagged = nltk.pos_tag(wordsList)
  
    print(POStagged)
```

# **6) Lemmatize the most frequently occurring 5 adjectives**

```python
from sys import float_info
#word_tokens = word_tokenize(text)
pos = nltk.pos_tag(flatList)
selective_pos = ['JJ','JJR', 'JJS']
selective_adj_words = []
for word,tag in pos:
     if tag in selective_pos:
         selective_adj_words.append((word,tag))
print(selective_adj_words)
```

```python
from collections import Counter

# Pass the split_it list to instance of Counter class.
Counters_found = Counter(selective_adj_words)
#print(Counters)

# most_common() produces k frequently encountered
# input values and their respective counts.
most_occur = Counters_found.most_common(5)
print(most_occur)
```
